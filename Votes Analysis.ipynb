## Task 2 : Votes Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ds = pd.read_csv(r'C:\Users\atewa\OneDrive\Desktop\Cognifyz\Dataset .csv')
```

### Identify the restaurants with the highest and lowest number of votes.

```python
ds.columns
```

```python
high_voted_res= ds.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending=False)
high_voted_res= high_voted_res.idxmax()
print("The Highest Rated Restaurant is : ", high_voted_res)
```

```python
low_voted_res= ds.groupby('Restaurant Name')['Votes'].sum().sort_values()
low_voted_res= low_voted_res.idxmin()
print("The Lowest Rated Restaurant is : ", low_voted_res)
```

### Analyze if there is a correlation between the number of votes and the rating of a restaurant.


```python
corr = np.corrcoef(ds['Votes'],ds['Aggregate rating'])
corr = np.round(corr,2)
print(corr)
```

```python
plt.figure(figsize=(9,5),dpi= 120)
sns.heatmap(corr, annot=True,fmt=".2f", linewidth=1)
plt.show()
```

```python
sns.scatterplot(data=ds , y='Votes', hue= 'Aggregate rating', x='Aggregate rating')
sns.regplot(data=ds, x='Aggregate rating', y='Votes', scatter=False)
plt.title(f"Correlation between Aggregate rating and Votes: 0.31")
plt.show()
print("Correlation value is 0.31 indicating there is a moderate positive correlation between Rating and Votes")
```
