## Task 3 : Price Range vs. Online Delivery and Table Booking

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ds = pd.read_csv(r'C:\Users\atewa\OneDrive\Desktop\Cognifyz\Dataset .csv')
```

### 1. Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

```python
ds.columns
```

### We need 2 charts to analyze this.

```python
x= ds['Has Online delivery']
y= ds['Has Table booking']
z= ds['Average Cost for two']
```

#### 1. Online delivery with the Average Cost
#### 2. Table Booking with the Average Cost

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x, y=z)
plt.title('Online Delivery affected by the Cost')
plt.xlabel('Online Delivery')
plt.ylabel('Average cost for Two')
plt.subplot(1,2,2)
sns.boxplot(x=y,y=z)
plt.title('Table Booking affected by the Cost')
plt.xlabel('Table Booking')
plt.ylabel('Average Cost for Two')
plt.tight_layout()

```

```python
print("As we can see, Both Online Delivery and Table Booking are not affected by Cost")
```

### 2. Determine if higher-priced restaurants are more likely to offer these services.


#### Checking if Restaurants with High Average cost more than 8,000 provide Online Delivery and Table booking.

```python
# we need box plot to compare these situations

plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x='Has Online delivery', y='Average Cost for two',data=ds.loc[ds['Average Cost for two']<8000], linewidth=1.5,color='red')
plt.title('Online Delivery by cost')
plt.subplot(1,2,2)
sns.boxplot(x='Has Table booking', y='Average Cost for two', data=ds.loc[ds['Average Cost for two']<8000], linewidth=1.5, color= 'green')
plt.title('Table Booking by Cost')
plt.tight_layout()
plt.show()
```

### Here, We can see that, Online delivery facility is available for only those restaurants, where the Average cost is less than 1000.
### And, The Table Booking facility is available for only those restaurants where Average cost is between 1000 and 2000.
