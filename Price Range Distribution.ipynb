## Task 3 : Price Range Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv(r'C:\Users\atewa\OneDrive\Desktop\Cognifyz\Dataset .csv')
```

### 1. Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
ds.columns
```

```python
ds.hist(column="Price range",color='skyblue')
plt.title('Distribution of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Counts')
plt.show()
```

### 2. Calculate the percentage of restaurants each price range category.

```python
price_range_cnt = ds["Price range"].value_counts()
total_res = len(ds)
percentage = round((price_range_cnt/total_res)*100,2)
percen_m= pd.merge(price_range_cnt, percentage, on="Price range")
print("The percentage of restaurants in each Price Range category are : ", percen_m)
```

```python
# Pie chart
plt.pie(percentage, labels=percentage.index, autopct='%1.1f%%', colors=["coral","#99ff99","yellow","#ffb3ba"])
plt.title("Price Range Percentage")
plt.show()
```
