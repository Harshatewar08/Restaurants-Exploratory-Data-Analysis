###  2. Calculate the average length of reviews and explore if there is a relationship between review length and rating.

```python
reviews = ds['Rating text']
total_reviews = len(reviews)
t_rev_length = 0

for review in reviews:
    t_rev_length += len(review.split())

avg_rev_length = t_rev_length/total_reviews

print('Average review length : ', round(avg_rev_length,2))
```

```python
rev_len_x_rating= {}

for rating in ds['Aggregate rating'].unique():
    fltrd_revs = ds.loc[ds['Aggregate rating']== rating]['Rating text']
    total_len = 0
    total_revs= len(fltrd_revs)

    for review in fltrd_revs:
        total_len += len(review.split())

    avg_len = total_len/total_revs
    rev_len_x_rating[rating] = avg_len

print('Relation between Rating and Review length : ')
print(rev_len_x_rating)
```

```python
rev_len_x_rating = pd.DataFrame({'rating': list(rev_len_x_rating.keys()), 'avg_review_len' : list(rev_len_x_rating.values()) })
rev_len_x_rating
```

```python
plt.figure(figsize=(13,5))
sns.barplot(x='rating', y='avg_review_len', data=rev_len_x_rating, )
plt.title('Ralation between Rating and Average Review Length ')
plt.xlabel('Rating ')
plt.ylabel('Average Revies Length')
plt.show()
```
